# Q2. 다음의 데이터는 연도, 국가별 기대수명을 나타내고 있는 데이터 입니다. 2011년 부터의 연도별, 국가별 평균 기대수명을 구해주세요.(2011년도 포함되게 구합니다.)

````python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
````

````python
df[df['Year']>=2011].groupby(['Year', 'Country'])['Life_Expectancy'].mean().unstack()
````
